```python
import pandas as pd
import re

def convert_table_to_excel(table_text, excel_filename):
    # Remove line numbers if present
    table_text = re.sub(r'^\s*\d+\|', '', table_text, flags=re.MULTILINE)
    # Remove any leading/trailing whitespace and split into lines
    lines = table_text.strip().split('\n')
    # Find the header line (the one with column names)
    header_line = None
    for i, line in enumerate(lines):
        if '|' in line and not set(line.strip()).issubset(set('-+| ')):
            header_line = i
            break
    if header_line is None:
        raise ValueError("Header line not found.")
    # Extract headers
    headers = [h.strip() for h in lines[header_line].split('|')]
    # Extract data lines (skip header and separator)
    data_lines = lines[header_line+2:]
    # Parse each data line
    data = []
    for line in data_lines:
        if not line.strip():
            continue
        row = [cell.strip() for cell in line.split('|')]
        if len(row) == len(headers):
            data.append(row)
    # Create DataFrame
    df = pd.DataFrame(data, columns=headers)
    # Write to Excel
    df.to_excel(excel_filename, index=False)
    print(f"Data written to {excel_filename}")

# Example usage:
table_text = """
     1| stock_code |           stock_name           | lot_size | expiry_date | strike_price |   cmp    |  ltp   |  margin   | right | start_of_strike 
     2|------------+--------------------------------+----------+-------------+--------------+----------+--------+-----------+-------+-----------------
     3| ABB        | ABB INDIA LIMITED              |      125 | 2025-08-28  |         3400 |   5090.5 |      0 |  90578.47 | call  |            5600
     4| ADAENT     | ADANI ENTERPRISES LIMITED      |      300 | 2025-08-28  |         2050 |     2188 | 170.75 |  14443.71 | call  |            2407
     5| ADAGAS     | ADANI TOTAL GAS LIMITED        |      775 | 2025-08-28  |          520 |      593 |     76 |  87278.23 | call  |             652
     6| ADAGRE     | ADANI GREEN ENERGY LTD         |      375 | 2025-08-28  |          440 |   923.35 |      0 |  82190.09 | call  |            1016
     7| ADAPOR     | ADANI PORT AND SPECIAL ECONO   |      400 | 2025-08-28  |         1040 |   1330.4 |      0 |  78473.82 | call  |            1463
     8| ADATRA     | ADANI ENERGY SOLUTIONS LIMITED |      625 | 2025-08-28  |          480 |   790.95 |      0 | 104755.17 | call  |             870
     9| ADICAP     | ADITYA BIRLA CAPITAL LIMITED   |     2700 | 2025-08-28  |          200 |   275.55 |      0 | 112924.42 | call  |             303
    10| ADIFAS     | ADITYA BIRLA FASHION & RTL LTD |     2600 | 2025-08-28  |           36 |     75.2 |      0 |  35246.81 | call  |              83
    11| ALKLAB     | ALKEM LABORATORIES LIMITED     |      100 | 2025-08-28  |         3300 |     4815 |      0 |  55408.08 | call  |            5297
    12| AMBCE      | AMBUJA CEMENTS LTD             |      900 | 2025-08-28  |          490 |    589.5 |      0 |   66785.7 | call  |             648
    13| ANGBRO     | ANGEL ONE LIMITED              |      200 | 2025-08-28  |         2800 |     2594 |   24.8 | 112867.04 | call  |            2853
    14| APLAPO     | APL APOLLO TUBES LIMITED       |      350 | 2025-08-28  |          680 |   1593.5 |      0 |   70332.8 | call  |            1753
    15| APOHOS     | APOLLO HOSPITALS ENTERPRISES L |      125 | 2025-08-28  |         5600 |   7131.5 |      0 |  97042.92 | call  |            7845
    16| ASHLEY     | ASHOK LEYLAND LTD              |     2500 | 2025-08-28  |           98 |    116.7 |      0 |  40178.71 | call  |             128
    17| ASIPAI     | ASIAN PAINTS INDIA LTD         |      200 | 2025-08-28  |         2600 |   2500.4 |  12.15 |  52415.48 | call  |            2750
    18| ASTPOL     | ASTRAL LIMITED                 |      367 | 2025-08-28  |          680 |   1384.6 |      0 |  61075.23 | call  |            1523
    19| AURPHA     | AUROBINDO PHARMA LTD           |      550 | 2025-08-28  |          600 |   1053.1 |      0 |  78270.19 | call  |            1158
    20| AUSMA      | AU SMALL FINANCE BANK LTD      |     1000 | 2025-08-28  |          670 |      755 |  62.25 |  85023.38 | call  |             831
    21| AVESUP     | AVENUE SUPERMARTS LTD DMART    |      150 | 2025-08-28  |         2100 |   4223.3 |      0 |  77313.99 | call  |            4646
    22| AXIBAN     | AXIS BANK LIMITED              |      625 | 2025-08-28  |          520 |   1060.2 |      0 |  74572.55 | call  |            1166
    23| BAAUTO     | BAJAJ AUTO LIMITED             |       75 | 2025-08-28  |         4400 |     8244 |      0 |  63308.44 | call  |            9068
    24| BAFINS     | BAJAJ FINSERV LIMITED          |      500 | 2025-08-28  |          960 |   1918.5 |      0 |  97995.82 | call  |            2110
    25| BAJFI      | BAJAJ FINANCE LIMITED          |      125 | 2025-08-28  |          100 |    880.2 |    790 |     11072 | call  |             968
    26| BANBAN     | BANDHAN BANK LIMITED           |     2800 | 2025-08-28  |           90 |   167.82 |      0 |  67685.65 | call  |             185
    27| BANBAR     | BANK OF BARODA                 |     2925 | 2025-08-28  |          130 |   240.34 |      0 |  92847.16 | call  |             264
    28| BANIND     | BANK OF INDIA                  |     4825 | 2025-08-28  |           50 |   111.27 |      0 |  79670.03 | call  |             122
    29| BHAAIR     | BHARTI AIRTEL LTD              |      475 | 2025-08-28  |         1040 |   1870.2 |      0 | 109812.01 | call  |            2057
    30| BHAELE     | BHARAT ELECTRONICS LTD         |     2850 | 2025-08-28  |          200 |    385.1 |      0 | 152560.78 | call  |             424
    31| BHAFOR     | BHARAT FORGE LIMITED           |      500 | 2025-08-28  |          520 |   1144.1 |      0 |  83645.12 | call  |            1259
    32| BHAINF     | INDUS TOWERS LIMITED           |     1700 | 2025-08-28  |          150 |    337.2 |      0 |  76257.37 | call  |             371
    33| BHAPET     | BHARAT PETROLEUM CORPN LTD     |     1800 | 2025-08-28  |          160 |    318.7 |      0 |  57946.82 | call  |             351
    34| BHEL       | BHARAT HEAVY ELECTRICALS LTD   |     2625 | 2025-08-28  |          110 |   225.36 |      0 | 106139.53 | call  |             248
    35| BIOCON     | BIOCON LIMITED                 |     2500 | 2025-08-28  |          200 |   354.95 |      0 |  140184.2 | call  |             390
    36| BOSLIM     | BOSCH LIMITED                  |       25 | 2025-08-28  |        54000 |    38595 |      0 | 103171.77 | call  |           42455
    37| BRIIND     | BRITANNIA INDUSTRIES LTD       |      100 | 2025-08-28  |         7100 |     5415 |      0 |  58669.26 | call  |            5957
    38| BSE        | BSE LIMITED                    |      125 | 2025-08-28  |         3000 |   2402.1 |    4.8 | 128776.83 | call  |            2642
    39| CADHEA     | ZYDUS LIFESCIENCES LIMITED     |      900 | 2025-08-28  |          650 |      940 |      0 |  90329.38 | call  |            1034
    40| CANBAN     | CANARA BANK                    |     6750 | 2025-08-28  |           78 |    108.2 |      0 |  98196.28 | call  |             119
    41| CDSL       | CENTRAL DEPOSITORY SER IND LTD |      350 | 2025-08-28  |         1060 |   1585.5 |      0 |  91962.78 | call  |            1744
    42|CESC       | CESC LIMITED                   |     2925 | 2025-08-28  |          203 |   164.42 |      0 |  73173.48 | call  |             181
    43| CHOINV     | CHOLAMANDALAM INVESTMENT AND F |      625 | 2025-08-28  |         1800 |   1469.9 |   0.55 | 132950.74 | call  |            1617
    44| CIPLA      | CIPLA LIMITED                  |      325 | 2025-08-28  |         1860 |   1488.4 |      0 |  50752.91 | call  |            1637
    45| CNXBAN     | NIFTY BANK                     |       30 | 2025-08-28  |        44900 |  55119.2 |      0 | 156707.59 | call  |           57875
    46| COALIN     | COAL INDIA LIMITED             |     1050 | 2025-08-28  |          345 |    379.9 |   30.5 |  40585.81 | call  |             418
    47| COLPAL     | COLGATE-PALMOLIVE (INDIA) LTD  |      175 | 2025-08-28  |         2000 |   2228.9 |    245 |  41276.39 | call  |            2452
    48| COMAGE     | COMPUTER AGE MANAGEMENT SERV   |      125 | 2025-08-28  |         2200 |   3804.5 |      0 |  76391.05 | call  |            4185
    49| CONCOR     | CONTAINER CORPORATION OF INDIA |     1000 | 2025-08-28  |          600 |   563.05 |    2.9 |  67253.42 | call  |             619
    50| CROGR      | CROMPTON GREAVES CONS ELEC LTD |     1800 | 2025-08-28  |          220 |    319.6 |      0 |  64576.51 | call  |             352
    51| CROGRE     | CG POWER AND IND SOLUTIONS LTD |      725 | 2025-08-28  |          420 |   676.35 |      0 |  75749.78 | call  |             744
    52| CUMIND     | CUMMINS INDIA LTD              |      150 | 2025-08-28  |         4800 |     3820 |      0 |  58780.77 | call  |            4202
    53| CYILIM     | CYIENT LIMITED                 |      300 | 2025-08-28  |         1840 |   1181.4 |      0 |   59693.4 | call  |            1300
    54| DABIND     | DABUR INDIA LTD                |     1250 | 2025-08-28  |          740 |      511 |      0 |  70021.12 | call  |             562
    55| DELLIM     | DELHIVERY LIMITED              |     1525 | 2025-08-28  |          495 |   464.85 |    3.9 |  103615.1 | call  |             511
    56| DIVLAB     | DIVIS LABORATORIES LIMITED     |      100 | 2025-08-28  |         9400 |   6019.5 |      0 |  69147.78 | call  |            6621
    57| DIXTEC     | DIXON TECHNOLOGIES INDIA LTD   |       50 | 2025-08-28  |        19250 |    16311 |      0 | 150248.54 | call  |           17942
    58| DLFLIM     | DLF LIMITED                    |      825 | 2025-08-28  |          860 |    756.5 |   1.15 |     82280 | call  |             832
    59| DRREDD     | DR REDDYS LABORATORIES LTD     |      625 | 2025-08-28  |         1340 |   1203.3 |   1.65 |  75061.02 | call  |            1324
    60| EICMOT     | EICHER MOTORS LTD              |      175 | 2025-08-28  |         6200 |     5693 |   5.55 | 100546.81 | call  |            6262
    61| EXIIND     | EXIDE INDUSTRIES LTD           |     1800 | 2025-08-28  |          275 |   376.45 |      0 |  89699.68 | call  |             414
    62| FEDBAN     | FEDERAL BANK LTD               |     5000 | 2025-08-28  |          150 |   197.63 |      0 | 106974.66 | call  |             217
    63| FSNECO     | FSN ECOMMERCE VENTURES NYKAA   |     2950 | 2025-08-28  |          225 |   205.74 |    1.5 |  87764.74 | call  |             226
    64| GAIL       | GAIL (INDIA) LIMITED           |     2350 | 2025-08-28  |          285 |   171.43 |      0 |  48220.24 | call  |             189
    65| GLEPHA     | GLENMARK PHARMACEUTICALS LTD   |      325 | 2025-08-28  |         1580 |   2031.6 |    575 |  99067.38 | call  |            2235
    66| GMRINF     | GMR AIRPORTS LIMITED           |     5625 | 2025-08-28  |          130 |    90.09 |      0 |  67988.73 | call  |              99
    67| GODCON     | GODREJ CONSUMER PRODUCTS LTD   |      500 | 2025-08-28  |          920 |   1207.1 |      0 |   65660.4 | call  |            1328
    68| GODPRO     | GODREJ PROPERTIES LIMITED      |      225 | 2025-08-28  |         2200 |   2010.1 |  11.75 |  75091.56 | call  |            2211
    69| GRANUL     | GRANULES INDIA LIMITED         |     1000 | 2025-08-28  |          740 |    445.6 |      0 |  138261.6 | call  |             490
    70| GRASIM     | GRASIM INDUSTRIES LIMITED      |      250 | 2025-08-28  |         2560 |   2699.9 |  156.3 |  76624.99 | call  |            2970
    71| HAVIND     | HAVELLS INDIA LIMITED          |      500 | 2025-08-28  |         1640 |   1498.8 |    2.4 |  76782.16 | call  |            1649
    72| HCLTEC     | HCL TECHNOLOGIES LIMITED       |      350 | 2025-08-28  |         1600 |   1472.1 |    1.8 |  54244.19 | call  |            1619
    73| HDFAMC     | HDFC ASSET MANAGEMENT CO LTD   |      150 | 2025-08-28  |         5200 |     5614 | 491.25 | 106560.61 | call  |            6175
    74| HDFBAN     | HDFC BANK LIMITED              |      550 | 2025-08-28  |         1500 |   1974.7 |      0 | 117757.58 | call  |            2172
    75| HDFSTA     | HDFC LIFE INSURANCE COM LTD    |     1100 | 2025-08-28  |          510 |   757.75 |      0 |  83456.23 | call  |             834
    76| HERHON     | HERO MOTOCORP LIMITED          |      150 | 2025-08-28  |         5950 |     4655 |      0 |   71908.9 | call  |            5121
    77| HIMFUT     | HFCL LIMITED                   |     4150 | 2025-08-28  |           96 |     73.6 |      0 |   6568.38 | call  |              81
    78| HINAER     | HINDUSTAN AERONAUTICS LIMITED  |      150 | 2025-08-28  |         2000 |     4470 |      0 | 102907.63 | call  |            4917
    79| HINDAL     | HINDALCO INDUSTRIES LIMITED    |     1400 | 2025-08-28  |          480 |    675.9 |      0 |  117598.2 | call  |             743
    80| HINLEV     | HINDUSTAN UNILEVER LIMITED     |      300 | 2025-08-28  |         2060 |   2506.6 |      0 |  80556.06 | call  |            2757
    81| HINPET     | HINDUSTAN PETROLEUM CORP       |     2025 | 2025-08-28  |          230 |    412.6 |      0 | 109528.31 | call  |             454
    82|HINZIN     | HINDUSTAN ZINC LTD             |     1225 | 2025-08-28  |          500 |    422.7 |    0.3 |  81028.32 | call  |             465
    83| HUDCO      | HOUSING URBAN DEV CO LTD HUDCO |     2425 | 2025-08-28  |          120 |   208.11 |      0 | 109084.31 | call  |             229
    84| ICIBAN     | ICICI BANK LIMITED             |      700 | 2025-08-28  |          940 |   1436.5 |      0 | 104694.04 | call  |            1580
    85| ICILOM     | ICICI LOMBARD GEN INSURANCE CO |      250 | 2025-08-28  |         2920 |   1880.5 |      0 |  52909.74 | call  |            2069
    86| ICIPRU     | ICICI PRUDENTIAL LIFE INS CO   |      750 | 2025-08-28  |          560 |    615.4 |     58 |  49954.71 | call  |             677
    87| IDECEL     | VODAFONE IDEA LIMITED          |    40000 | 2025-08-28  |           26 |     6.61 |      0 |  59983.36 | call  |               7
    88| IDFBAN     | IDFC FIRST BANK LIMITED        |     7500 | 2025-08-28  |           53 |    69.39 |      0 |  58879.47 | call  |              76
    89| IIFHOL     | IIFL FINANCE LIMITED           |     1550 | 2025-08-28  |          740 |   454.15 |      0 | 140863.51 | call  |             500
    90| INDBA      | INDUSIND BANK LIMITED          |      500 | 2025-08-28  |          620 |   794.65 |      0 |  75934.17 | call  |             874
    91| INDEN      | INDIAN ENERGY EXCHANGE LIMITED |     3750 | 2025-08-28  |          103 |    138.9 |      0 |  98871.37 | call  |             153
    92| INDGAS     | INDRAPRASHTHA GAS LTD          |     2750 | 2025-08-28  |          240 |    203.7 |   0.25 |  91094.66 | call  |             224
    93| INDHOT     | INDIAN HOTELS CO LTD           |     1000 | 2025-08-28  |          900 |    739.8 |    0.2 |  92352.52 | call  |             814
    94| INDIBA     | INDIAN BANK                    |      950 | 2025-08-28  |          320 |    654.7 |      0 |  78353.36 | call  |             720
    95| INDOIL     | INDIAN OIL CORPORATION LIMITED |     4875 | 2025-08-28  |          117 |   141.35 |   20.6 |  68943.88 | call  |             155
    96| INDR       | INDIAN RAILWAY FIN CORP LTD    |     3525 | 2025-08-28  |           75 |   126.31 |      0 |  83378.84 | call  |             139
    97| INDRAI     | INDIAN RAIL CAT AND TOUR IRCTC |      875 | 2025-08-28  |          360 |   723.35 |      0 |  71981.74 | call  |             796
    98| INDREN     | INDIAN RENEW. ENG.DEV. AGY LTD |     2900 | 2025-08-28  |           80 |    143.6 |      0 |  84357.36 | call  |             158
    99| INFEDG     | INFO EDGE (INDIA) LIMITED      |       75 | 2025-08-28  |          720 |   1351.6 |      0 |  14159.04 | call  |            1487
   100| INFTEC     | INFOSYS LTD                    |      400 | 2025-08-28  |          720 |   1423.1 |      0 |  60553.58 | call  |            1565
   101| INOWIN     | INOX WIND LIMITED              |     3225 | 2025-08-28  |          177 |   143.25 |   0.25 | 112048.41 | call  |             158
   102| INTAVI     | INTERGLOBE AVIATION LIMITED    |      150 | 2025-08-28  |         3400 |   5820.5 |      0 |  95849.26 | call  |            6403
   103| IRBINF     | IRB INFRASTRUCTURE DEVELOPERS  |    10250 | 2025-08-28  |           76 |    44.26 |      0 |  83659.68 | call  |              49
   104| ITC        | ITC LIMITED                    |     1600 | 2025-08-28  |          515 |   413.85 |      0 |  66584.54 | call  |             455
   105| JINSP      | JINDAL STEEL & POWER LIMITED   |      625 | 2025-08-28  |          520 |   978.65 |      0 |  87920.88 | call  |            1077
   106| JINSTA     | JINDAL STAINLESS LIMITED       |      775 | 2025-08-28  |          930 |   692.95 |      0 | 102113.46 | call  |             762
   107| JIOFIN     | JIO FINANCIAL SERVICES LIMITED |     1650 | 2025-08-28  |          190 |      323 |      0 |  69537.13 | call  |             355
   108| JSWENE     | JSW ENERGY LIMITED             |      750 | 2025-08-28  |          240 |   529.15 |      0 |  67097.29 | call  |             582
   109| JSWSTE     | JSW STEEL LIMITED              |      675 | 2025-08-28  |          480 |   1052.4 |      0 |  80820.11 | call  |            1158
   110| JUBFOO     | JUBILANT FOODWORKS LIMITED     |     1250 | 2025-08-28  |          340 |    633.7 |      0 |  96108.35 | call  |             697
   111| KALJEW     | KALYAN JEWELLERS INDIA LIMITED |      775 | 2025-08-28  |          280 |   544.25 |      0 | 119467.67 | call  |             599
   112| KEIIND     | KEI INDUSTRIES LIMITED         |      150 | 2025-08-28  |         3300 |   3859.4 |    682 | 113218.92 | call  |            4245
   113| KOTMAH     | KOTAK MAHINDRA BANK LTD        |      400 | 2025-08-28  |         2420 |   1970.3 |      0 |  84890.41 | call  |            2167
   114| KPITE      | KPIT TECHNOLOGIES LIMITED      |      400 | 2025-08-28  |         1180 |   1200.7 |  39.75 |  75163.92 | call  |            1321
   115| LARTOU     | LARSEN AND TOUBRO LIMITED      |      150 | 2025-08-28  |         4250 |   3636.9 |      0 |  11689.53 | call  |            4001
   116| LAULAB     | LAURUS LABS LIMITED            |     1700 | 2025-08-28  |          980 |   831.25 |    0.9 | 213430.75 | call  |             914
   117| LIC        | LIFE INSURANCE CORP OF IND LIC |      575 | 2025-08-28  |          520 |    916.7 |      0 |  51359.19 | call  |            1008
   118| LICHF      | LIC HOUSING FINANCE LIMITED    |     1000 | 2025-08-28  |          740 |    573.8 |   0.15 | 100757.36 | call  |             631
   119| LTFINA     | L&T FINANCE LIMITED            |     4462 | 2025-08-28  |          245 |   198.42 |   0.05 | 122739.89 | call  |             218
   120| LTINFO     | LTIMINDTREE LIMITED            |      150 | 2025-08-28  |         2600 |     5040 |      0 |  83690.88 | call  |            5544
   121| LUPIN      | LUPIN LIMITED                  |      425 | 2025-08-28  |         1660 |   1930.5 | 286.65 |  93322.55 | call  |            2124
   122|MACDEV     | MACROTECH DEVELOPERS LIMITED   |      450 | 2025-08-28  |         1060 |   1213.1 |      0 | 101936.95 | call  |            1334
   123| MAGFI      | POONAWALLA FINCORP LIMITED     |     1450 | 2025-08-28  |          310 |   440.25 |      0 | 103595.83 | call  |             484
   124| MAHMAH     | MAHINDRA & MAHINDRA LIMITED    |      175 | 2025-08-28  |         2800 |   3191.5 |    421 |  61508.18 | call  |            3511
   125| MANAFI     | MANAPPURAM FINANCE LIMITED     |     3000 | 2025-08-28  |          215 |   261.95 |      0 | 120083.34 | call  |             288
   126| MARLIM     | MARICO LIMITED                 |     1200 | 2025-08-28  |          650 |    712.6 |   65.5 |  90691.22 | call  |             784
   127| MARUTI     | MARUTI SUZUKI INDIA LTD        |       50 | 2025-08-28  |        13300 |    12691 |  35.05 |  64278.24 | call  |           13960
   128| MAXFIN     | MAX FINANCIAL SERVICES LIMITED |      800 | 2025-08-28  |         1460 |   1537.3 |     97 |  115876.8 | call  |            1691
   129| MAXHEA     | MAX HEALTHCARE INSTITUTE LTD   |      525 | 2025-08-28  |         1880 |   1252.6 |      0 |  103017.1 | call  |            1378
   130| MCX        | MULTI COMMODITY EXCHANGE       |      100 | 2025-08-28  |         4200 |   7766.5 |      0 |  140215.4 | call  |            8543
   131| MOTSUM     | SAMVARDHANA MOTHERSON INT LTD  |     3550 | 2025-08-28  |          125 |    93.39 |      0 |  49931.34 | call  |             103
   132| MPHLIM     | MPHASIS LIMITED                |      275 | 2025-08-28  |         2593 |     2684 |    240 |  97081.38 | call  |            2952
   133| MUTFIN     | MUTHOOT FINANCE LIMITED        |      275 | 2025-08-28  |         1880 |     2657 |      0 |  81105.83 | call  |            2923
   134| NAGCON     | NCC LIMITED                    |     1775 | 2025-08-28  |          185 |    224.9 |   36.5 |  66089.82 | call  |             247
   135| NATALU     | NATIONAL ALUMINIUM CO LTD      |     3750 | 2025-08-28  |          185 |   187.17 |    6.4 | 125196.24 | call  |             206
   136| NATMIN     | NMDC LIMITED                   |    13500 | 2025-08-28  |           55 |    71.13 |      0 |  137205.9 | call  |              78
   137| NBCC       | NBCC (INDIA) LIMITED           |     5800 | 2025-08-28  |           78 |   107.26 |      0 |  13608.51 | call  |             118
   138| NESIND     | NESTLE INDIA LIMITED           |      200 | 2025-08-28  |         1000 |   1113.9 |    119 |  23496.82 | call  |            1225
   139| NHPC       | NHPC LIMITED                   |     6400 | 2025-08-28  |           82 |    82.33 |   2.45 |  81465.11 | call  |              91
   140| NIFFIN     | NIFTY FINANCIAL SERVICES INDEX |       65 | 2025-08-28  |        20400 | 26201.85 |      0 |  160856.1 | call  |           27512
   141| NIFNEX     | NIFTY NEXT 50                  |       25 | 2025-08-28  |        56100 |  66040.6 |      0 | 146504.85 | call  |           69343
   142| NIFSEL     | NIFTY MIDCAP SELECT            |      120 | 2025-08-28  |        15725 | 12616.05 |      0 | 140954.38 | call  |           13247
   143| NIFTY      | NIFTY 50                       |       75 | 2025-08-28  |        19300 | 24431.35 |      0 | 136773.96 | call  |           25653
   144| NIITEC     | COFORGE LIMITED                |       75 | 2025-08-28  |         1320 |     1633 |      0 |  23712.79 | call  |            1796
   145| NTPC       | NTPC LIMITED                   |     1500 | 2025-08-28  |          470 |    332.8 |      0 |  48091.68 | call  |             366
   146| OBEREA     | OBEROI REALTY LIMITED          |      350 | 2025-08-28  |          920 |   1589.7 |      0 |  81243.89 | call  |            1749
   147| ODICEM     | DALMIA BHARAT LIMITED          |      275 | 2025-08-28  |         1240 |     2256 |      0 |  67341.13 | call  |            2482
   148| OILIND     | OIL INDIA LIMITED              |     1075 | 2025-08-28  |          180 |   434.95 |      0 |  75837.31 | call  |             478
   149| ONE97      | ONE 97 COMMUNICATIONS PAYTM    |      650 | 2025-08-28  |          520 |   1067.9 |      0 | 171575.75 | call  |            1175
   150| ONGC       | OIL AND NATURAL GAS CORPORATIO |     1925 | 2025-08-28  |          110 |   233.88 |      0 |  50285.59 | call  |             257
   151| ORAFIN     | ORACLE FINANCIAL SERVICES SOFT |       50 | 2025-08-28  |         4200 |   8516.5 |      0 |  66579.89 | call  |            9368
   152| PAGIND     | PAGE INDUSTRIES LTD            |       15 | 2025-08-28  |        25000 |    44670 |      0 |  80556.06 | call  |           49137
   153| PBFINT     | PB FINTECH LTD POLICYBZR       |      325 | 2025-08-28  |          800 |   1773.7 |      0 |  12337.47 | call  |            1951
   154| PERSYS     | PERSISTENT SYSTEMS LIMITED     |      100 | 2025-08-28  |         2600 |     5146 |      0 |   78061.1 | call  |            5661
   155| PETLNG     | PETRONET LNG LIMITED           |     1500 | 2025-08-28  |          120 |    275.9 |      0 |   49046.4 | call  |             303
   156| PHOMIL     | PHOENIX MILLS LTD              |      350 | 2025-08-28  |          600 |   1442.4 |      0 |  84502.39 | call  |            1587
   157| PIDIND     | PIDILITE INDUSTRIES LTD        |      250 | 2025-08-28  |         1520 |   3042.9 |      0 |   80048.8 | call  |            3347
   158| PIIND      | P I INDUSTRIES LTD             |      125 | 2025-08-28  |         2200 |     3968 |      0 |  57274.75 | call  |            4365
   159| PNBHOU     | PNB HOUSING FINANCE LIMITED    |      650 | 2025-08-28  |          280 |   768.55 |      0 | 119272.99 | call  |             845
   160| POLI       | POLYCAB INDIA LIMITED          |      125 | 2025-08-28  |         3400 |   6853.5 |      0 | 138244.47 | call  |            7539
   161| POWFIN     | POWER FINANCE CORPORATION LTD  |     1300 | 2025-08-28  |          220 |    406.5 |      0 |  89844.79 | call  |             447
   162| POWGRI     | POWER GRID CORPORATION OF INDI |     1800 | 2025-08-28  |          160 |    284.4 |      0 |  52948.11 | call  |             313
   163|PREEST     | PRESTIGE ESTATES PROJECTS LIMI |      325 | 2025-08-28  |          880 |   1617.9 |      0 |  97683.36 | call  |            1780
   164| PUNBAN     | PUNJAB NATIONAL BANK           |     8000 | 2025-08-28  |           56 |   104.23 |      0 | 114268.12 | call  |             115
   165| RBLBAN     | RBL BANK LIMITED               |     2500 | 2025-08-28  |          150 |    269.8 |      0 |  213346.8 | call  |             297
   166| RELIND     | RELIANCE INDUSTRIES            |      500 | 2025-08-28  |          720 |   1376.8 |      0 |  74572.68 | call  |            1514
   167| RUCSOY     | PATANJALI FOODS LIMITED        |      300 | 2025-08-28  |          880 |   1812.2 |      0 |  69165.24 | call  |            1993
   168| RURELE     | REC LIMITED                    |     1000 | 2025-08-28  |          160 |    383.6 |      0 |  62580.18 | call  |             422
   169| SAIL       | STEEL AUTHORITY OF INDIA LTD   |     4000 | 2025-08-28  |           60 |   121.59 |      0 | 159078.01 | call  |             134
   170| SBICAR     | SBI CARDS AND PAYMENT SERV LTD |      800 | 2025-08-28  |          320 |    789.6 |      0 |  70410.08 | call  |             869
   171| SBILIF     | SBI LIFE INSURANCE COMPANY LTD |      375 | 2025-08-28  |          960 |   1844.3 |      0 |  73647.27 | call  |            2029
   172| SHRCEM     | SHREE CEMENTS LTD              |       25 | 2025-08-28  |        16500 |    30635 |      0 |  81085.81 | call  |           33699
   173| SHRTRA     | SHRIRAM FINANCE LIMITED        |      750 | 2025-08-28  |          320 |   616.05 |      0 |  75018.25 | call  |             678
   174| SIEMEN     | SIEMENS LTD                    |       75 | 2025-08-28  |          800 |   3101.5 |      0 |  33294.43 | call  |            3412
   175| SJVLIM     | SATLUJ JAL VIDYUT NIGAM LTD    |     4725 | 2025-08-28  |           45 |    93.11 |      0 |   88972.4 | call  |             102
   176| SOLIN      | SOLAR INDUSTRIES INDIA LIMITED |       75 | 2025-08-28  |         6500 |    14465 |      0 | 191633.78 | call  |           15912
   177| SONBLW     | SONA BLW PRECISION FORGINGS    |      775 | 2025-08-28  |          680 |   446.45 |      0 |  49407.33 | call  |             491
   178| SRF        | SRF LIMITED                    |      375 | 2025-08-28  |         1500 |     2878 |      0 | 124142.52 | call  |            3166
   179| STABAN     | STATE BANK OF INDIA            |      750 | 2025-08-28  |          440 |   800.55 |      0 |  62197.91 | call  |             881
   180| SUNPHA     | SUN PHARMACEUTICAL INDUSTRIES  |      350 | 2025-08-28  |          880 |   1586.9 |      0 |  59045.89 | call  |            1746
   181| SUPIND     | SUPREME INDUSTRIES LTD         |      125 | 2025-08-28  |         2200 |     4135 |      0 |   78223.6 | call  |            4549
   182| SYNINT     | SYNGENE INTERNATIONAL LTD      |     1000 | 2025-08-28  |          280 |   666.95 |      0 |  78085.28 | call  |             734
   183| TATCHE     | TATA CHEMICALS LTD             |      550 | 2025-08-28  |          480 |   951.95 |      0 |  67046.98 | call  |            1047
   184| TATELX     | TATA ELXSI LIMITED             |      100 | 2025-08-28  |         2800 |   5818.5 |      0 |  72530.12 | call  |            6400
   185| TATGLO     | TATA CONSUMER PRODUCTS LIMITED |      456 | 2025-08-28  |          580 |   1056.1 |      0 |  48969.63 | call  |            1162
   186| TATMOT     | TATA MOTORS LIMITED            |      550 | 2025-08-28  |          340 |    641.9 |      0 |  44706.61 | call  |             706
   187| TATPOW     | TATA POWER CO LTD              |     1350 | 2025-08-28  |          200 |    382.5 |      0 |  62540.12 | call  |             421
   188| TATSTE     | TATA STEEL LIMITED             |     5500 | 2025-08-28  |           80 |    157.7 |      0 | 106268.73 | call  |             173
   189| TATTEC     | TATA TECHNOLOGIES LIMITED      |      700 | 2025-08-28  |          300 |   665.15 |      0 |  49014.05 | call  |             732
   190| TCS        | TATA CONSULTANCY SERVICES LTD  |      175 | 2025-08-28  |         1560 |   3050.9 |      0 |  55634.57 | call  |            3356
   191| TECMAH     | TECH MAHINDRA LIMITED          |      600 | 2025-08-28  |          720 |   1490.4 |      0 |  91175.13 | call  |            1639
   192| TITIND     | TITAN COMPANY LIMITED          |      175 | 2025-08-28  |         3040 |   3475.9 |      0 |  13021.36 | call  |            3823
   193| TITWAG     | TITAGARH RAIL SYSTEMS LIMITED  |      625 | 2025-08-28  |          520 |      828 |      0 | 125331.86 | call  |             911
   194| TORPHA     | TORRENT PHARMACEUTICALS LTD    |      250 | 2025-08-28  |         1700 |   3593.7 |      0 |  90910.69 | call  |            3953
   195| TORPOW     | TORRENT POWER LIMITED          |      375 | 2025-08-28  |          680 |     1340 |      0 |  88499.51 | call  |            1474
   196| TRENT      | TRENT LTD                      |      100 | 2025-08-28  |         2600 |   5320.5 |      0 |   86583.9 | call  |            5853
   197| TUBIN      | TUBE INVESTMENTS OF INDIA LTD  |      150 | 2025-08-28  |         1600 |   2909.1 |      0 |  66949.19 | call  |            3200
   198| TVSMOT     | TVS MOTOR COMPANY LIMITED      |      350 | 2025-08-28  |         1600 |   2993.8 |      0 | 115587.83 | call  |            3293
   199| ULTCEM     | ULTRATECH CEMENT LIMITED       |       50 | 2025-08-28  |         6400 |    12210 |      0 |   65149.5 | call  |           13431
   200| UNIBAN     | UNION BANK OF INDIA            |     4425 | 2025-08-28  |          135 |   131.45 |    2.1 |   94919.7 | call  |             145
   201| UNIP       | UPL LIMITED                    |     1355 | 2025-08-28  |          360 |   694.65 |      0 | 120163.38 | call  |             764
   202| UNISPI     | UNITED SPIRITS LIMITED         |      350 | 2025-08-28  |          680 |   1305.6 |      0 |   45240.1 | call  |            1436
   203|VARBEV     | VARUN BEVERAGES LIMITED        |      875 | 2025-08-28  |          200 |    500.2 |      0 |  56906.91 | call  |             550
   204| VEDLIM     | VEDANTA LIMITED                |     1150 | 2025-08-28  |          200 |   433.95 |      0 |  70005.46 | call  |             477
   205| VOLTAS     | VOLTAS LTD                     |      300 | 2025-08-28  |          960 |   1303.7 |      0 |  53205.98 | call  |            1434
   206| WIPRO      | WIPRO LTD                      |     3000 | 2025-08-28  |          248 |   240.45 |    2.5 |  80312.54 | call  |             264
   207| YESBAN     | YES BANK LIMITED               |    26000 | 2025-08-28  |            2 |    18.69 |      0 |  70762.32 | call  |              21
   208| ZOMLIM     | ETERNAL LIMITED                |     2000 | 2025-08-28  |          325 |    301.4 |   1.95 |    102089 | call  |             332
"""
convert_table_to_excel(table_text, "output.xlsx")
```
